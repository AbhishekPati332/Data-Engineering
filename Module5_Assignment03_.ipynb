```python
#6. Explain the concept of generators in Python and how they are defined
```

```python
A generator in Python is a type of iterator that allows to generate values on the fly,
instead of storing them all in memory at once.
Generator are used to handle large data efficiently.
Generators are defined using functions with the yield keyword instead of return.
```

```python
#Ex:-
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
```

```python
gen = count_up_to(3)
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
#the list.

def sum_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

# Example list
my_list = [1, 2, 3, 4, 5, 6]

# Calling the function and print the result
print(sum_even_numbers(my_list))
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number ofterms.
```

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_value = self.a + self.b
            self.a, self.b = self.b, next_value
            self.count += 1
            return next_value

# Example usage
n = int(input("Enter the number of Fibonacci terms: "))
fib = FibonacciIterator(n)

print("Fibonacci sequence:")
for num in fib:
    print(num, end=' ')
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
```

```python
def powers_of_two(max_exponent):
  for i in range(max_exponent + 1):
    yield 2 ** i

n = int(input("Enter the maximum exponent: "))
print("Powers of 2:")
for power in powers_of_two(n):
  print(power, end='')


#we want this thing:- 2^0, 2^1, 2^2, ..., 2^n
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
```

```python
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()  # strip() removes newline characters

file_name = input("Enter the file name: ")
for line in read_file_line_by_line(file_name):
    print(line)
```

```python
def c_to_f(celsius):
    return (celsius * 9/5) + 32

celsius_list = [0, 20, 37, 100]

fahrenheit_list = list(map(c_to_f, celsius_list))

# Print the result
print("Temperatures in Fahrenheit:")
print(fahrenheit_list)
```

```python
#10. Create a Python program that uses filter() to remove all the vowels from a given string.
```

```python
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

text = input("Enter a string: ")

# Use filter to remove vowels
no_vowels = ''.join(filter(is_not_vowel, text))

print("String without vowels:", no_vowels)
```
